## Scaling

```python
# input feature and target
feature_columns = ['date','log_strike_price','maturity','t30ret','log_spindx','impl_volatility', 'Vola_ATM']
target_column = 'log_price'
```

```python
surface_features = surface_df[feature_columns]
surface_targets = surface_df[target_column]
```

```python
from sklearn.model_selection import train_test_split

# train test split

X_train, X_test, y_train, y_test= train_test_split(surface_features, surface_targets, test_size=0.3, random_state= 42, shuffle=True)

X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size=0.5, random_state= 42, shuffle=True)

print('Train: {}, Test: {}, Val: {}'.format(len(X_train), len(X_test), len(X_val)))
```

```python
from sklearn.preprocessing import RobustScaler

scaling_columns= ['log_strike_price','maturity','t30ret','log_spindx','impl_volatility', 'Vola_ATM']

# Initialize the RobustScaler
scaler_input = RobustScaler()

# Fit the scaler to the training data only
scaler_input.fit(X_train[scaling_columns])

# Transform the training data
X_train_scaled = scaler_input.transform(X_train[scaling_columns])

# Transform the validation data using the same scaler
X_val_scaled = scaler_input.transform(X_val[scaling_columns])

# Transform the test data using the same scaler
X_test_scaled = scaler_input.transform(X_test[scaling_columns])
```

```python
from sklearn.preprocessing import RobustScaler

# Initialize the RobustScaler
scaler_target = RobustScaler()

# Fit the scaler to the training data only
scaler_target.fit(y_train.values.reshape(-1, 1))

# Transform the training data
y_train_scaled = scaler_target.transform(y_train.values.reshape(-1, 1))

# Transform the validation data using the same scaler
y_val_scaled = scaler_target.transform(y_val.values.reshape(-1, 1))

# Transform the test data using the same scaler
y_test_scaled = scaler_target.transform(y_test.values.reshape(-1, 1))
```

```python
import pandas as pd

# Extract date columns from the original df
date_train = X_train[['date']].reset_index(drop=True)
date_val = X_val[['date']].reset_index(drop=True)
date_test = X_test[['date']].reset_index(drop=True)

# Convert scaled arrays back to pandas df for the input features
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=scaling_columns)
X_val_scaled_df = pd.DataFrame(X_val_scaled, columns=scaling_columns)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=scaling_columns)

# Add date column back to the scaled df
X_train_scaled_df = pd.concat([date_train, X_train_scaled_df], axis=1)
X_val_scaled_df = pd.concat([date_val, X_val_scaled_df], axis=1)
X_test_scaled_df = pd.concat([date_test, X_test_scaled_df], axis=1)

# Convert scaled targets back to Series
y_train_scaled_series = pd.Series(y_train_scaled.flatten(), name='log_price')
y_val_scaled_series = pd.Series(y_val_scaled.flatten(), name='log_price')
y_test_scaled_series = pd.Series(y_test_scaled.flatten(), name='log_price')

# Combine scaled inputs and targets into single df
train_scaled_df = pd.concat([X_train_scaled_df, y_train_scaled_series], axis=1)
val_scaled_df = pd.concat([X_val_scaled_df, y_val_scaled_series], axis=1)
test_scaled_df = pd.concat([X_test_scaled_df, y_test_scaled_series], axis=1)

# Concatenate all three df into one, reset index
combined_scaled_df = pd.concat([train_scaled_df, val_scaled_df, test_scaled_df], axis=0).reset_index(drop=True)
```

```python
# datetime
train_scaled_df['date'] = pd.to_datetime(train_scaled_df['date'])
val_scaled_df['date'] = pd.to_datetime(val_scaled_df['date'])
test_scaled_df['date'] = pd.to_datetime(test_scaled_df['date'])
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import os
import pandas as pd
from scipy.interpolate import griddata
import numpy as np

def plot_volatility_surface(plot_dataframe, file_name):
    # where plots are saved
    plots_dir = os.path.join("/content/drive/MyDrive/Colab Notebooks/BAThesis", file_name)
    os.makedirs(plots_dir, exist_ok=True)

    # ensure 'date' column is in datetime format
    plot_dataframe['date'] = pd.to_datetime(plot_dataframe['date'])

    # get min and max values for each axis
    x_min, x_max = np.min(plot_dataframe['log_strike_price']), np.max(plot_dataframe['log_strike_price'])
    y_min, y_max = np.min(plot_dataframe['maturity']), np.max(plot_dataframe['maturity'])
    z_min, z_max = np.min(plot_dataframe['impl_volatility']), np.max(plot_dataframe['impl_volatility'])

    # iterate through each trading day
    for current_date in plot_dataframe['date'].unique():
        df_final_day = plot_dataframe[plot_dataframe['date'] == current_date]
        df_grouped = df_final_day.groupby(['log_strike_price', 'maturity'])['impl_volatility'].mean().reset_index()

        # create unique grids for both strike prices and maturities
        strike_price_unique = np.unique(df_grouped['log_strike_price'])
        maturity_unique = np.unique(df_grouped['maturity'])

        # create a meshgrid for strike prices and maturities
        X, Y = np.meshgrid(strike_price_unique, maturity_unique)

        # interpolate Z values = implied volatility on this grid
        Z = griddata((df_grouped['log_strike_price'], df_grouped['maturity']), df_grouped['impl_volatility'], (X, Y), method='cubic')

        # plotting
        plt.style.use('seaborn-v0_8-whitegrid')
        fig = plt.figure(figsize=(12, 10))
        ax = fig.add_subplot(111, projection='3d')

        # using plot_surface to plot the volatility surface
        ax.plot_surface(X, Y, Z, cmap='inferno', vmax=z_max, vmin=z_min, linewidth=0, antialiased=False)

        title_date = pd.to_datetime(current_date).date()

        # adding title and labels
        ax.set_title(f'Volatility Surface, {title_date}', fontsize=18, pad=20)
        ax.set_xlabel('Strike Price', fontsize=14, labelpad=15)
        ax.set_ylabel('Time to Maturity', fontsize=14, labelpad=15)
        ax.set_zlabel('Implied Volatility', fontsize=14, labelpad=15)

        # set fixed limits for each axis
        ax.set_xlim(x_min, x_max)
        ax.set_ylim(y_min, y_max)
        ax.set_zlim(z_min, z_max)

        # show gridlines and set the background color
        ax.grid(True, linestyle='--', color='black', linewidth=0.5)

        plt.tight_layout()

        # save the plot
        filename = os.path.join(plots_dir, f"vol_surface_{title_date}.png")
        plt.savefig(filename, dpi=100, format='jpeg')
        plt.close(fig)

    # close all plots
    plt.close('all')
```

```python
import torch
from torch import nn, optim
from torch.utils.data import DataLoader
from sklearn.metrics import mean_squared_error
from tqdm import tqdm
import pandas as pd
import numpy as np
import copy

def training_function(model_constructor, train_dataset, val_dataset, device, learning_rates, num_epochs, batch_sizes):
    """
    Trains and validates a given model with specified hyperparameters.

    Parameters:
    - model_constructor: The neural network model to train.
    - train_dataset: Training dataset.
    - val_dataset: Validation dataset.
    - device: The device to run the training on ('cuda' or 'cpu').
    - learning_rates: List of learning rates to use for training.
    - num_epochs: Number of epochs to train for each learning rate.
    - batch_sizes: List of batch sizes to use for training.

    Returns:
    - A DataFrame containing training and validation losses and RMSE for each combination of hyperparameters.
    """
    results = []
    best_val_rmse = float('inf')
    best_model = None

    for lr in learning_rates:
        for batch_size in batch_sizes:

            # Initialize DataLoader for current batch size
            train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=False)
            val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

            # Reset model and optimizer for each learning rate
            model = model_constructor()
            model.to(device)
            optimizer = optim.Adam(model.parameters(), lr=lr)
            criterion = nn.MSELoss()

            for epoch in tqdm(range(num_epochs), desc=f"LR={lr}, Batch={batch_size}"):
                # Training phase
                model.train()
                train_loss = 0
                for (images, additional_features), targets in train_loader:

                    images, additional_features, targets = images.to(device), additional_features.to(device), targets.to(device)
                    optimizer.zero_grad()
                    outputs = model((images, additional_features))
                    loss = criterion(outputs.squeeze(1), targets)
                    loss.backward()
                    optimizer.step()
                    train_loss += loss.item()

                avg_train_loss = train_loss / len(train_loader)

                # Validation phase
                model.eval()
                val_loss = 0
                predictions, true_values = [], []
                with torch.no_grad():
                    for (images, additional_features), targets in val_loader:
                        images, additional_features, targets = images.to(device), additional_features.to(device), targets.to(device)
                        outputs = model((images, additional_features))
                        loss = criterion(outputs.squeeze(1), targets)
                        val_loss += loss.item()
                        predictions.extend(outputs.squeeze(1).cpu().numpy())
                        true_values.extend(targets.cpu().numpy())

                avg_val_loss = val_loss / len(val_loader)
                rmse = np.sqrt(mean_squared_error(true_values, predictions))

                if rmse < best_val_rmse:
                    best_val_rmse = rmse
                    best_model = copy.deepcopy(model)

                results.append({
                    ('Hyperparameters', 'Learning Rate'): lr,
                    ('Hyperparameters', 'Batch Size'): batch_size,
                    ('Metrics', 'Epoch'): epoch + 1,
                    ('Metrics', 'Training Loss'): avg_train_loss,
                    ('Metrics', 'Validation Loss'): avg_val_loss,
                    ('Metrics', 'RMSE'): rmse
                })

    return pd.DataFrame(results), model, best_model
```

## Evalutation

```python
def evaluate_model_on_test_set(model, test_dataset, batch, device):
    """
    Evaluates the given model on the test dataset.

    Parameters:
    - model: The trained neural network model.
    - test_dataset: Test dataset.
    - device: The device to run the evaluation on ('cuda' or 'cpu').

    Returns:
    - A tuple of (actual_prices, predicted_prices) as numpy arrays.
    """
    model.eval()
    test_loader = DataLoader(test_dataset, batch_size=batch, shuffle=False)
    model.to(device)

    actual_prices = []
    predicted_prices = []

    with torch.no_grad():
        for (images, additional_features), targets in test_loader:
            images, additional_features, targets = images.to(device), additional_features.to(device), targets.to(device)
            outputs = model((images, additional_features))
            actual_prices.extend(targets.cpu().numpy())
            predicted_prices.extend(outputs.squeeze(1).cpu().numpy())

    # Reshape for scaler if necessary
    actual_prices_reshaped = np.array(actual_prices).reshape(-1, 1)
    predicted_prices_reshaped = np.array(predicted_prices).reshape(-1, 1)

    # Inverse transform to get predictions and actuals in original scale
    true_price_predictions = scaler_target.inverse_transform(predicted_prices_reshaped).flatten()
    true_actual_prices = scaler_target.inverse_transform(actual_prices_reshaped).flatten()

    # transform back from log
    true_price_predictions = np.exp(true_price_predictions)
    true_actual_prices = np.exp(true_actual_prices)

    return true_actual_prices, true_price_predictions
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def evaluate_model_and_display_results(model, datasets, batch_size, device):
    """
    evaluate model on multiple datasets and display result

    Parameters:
    - model: trained model
    - datasets: dictionary containing datasets for evaluation
    - batch_size: chosen batch size for model
    - device: cuda, cpu

    Returns:
    - tuple containing df with results of actual and predicted prices + dictionary with metrics results for each period
    """

    all_results = []
    metrics_results = {}

    # evaluate  model dataset
    for period, dataset in datasets.items():
        actual_prices, predicted_prices = evaluate_model_on_test_set(model=model, test_dataset=dataset, batch=batch_size, device=device)

        # calculate  metrics and append to dict
        mse = mean_squared_error(actual_prices, predicted_prices)
        rmse = np.sqrt(mse)
        mae = mean_absolute_error(actual_prices, predicted_prices)
        r2 = r2_score(actual_prices, predicted_prices)
        metrics_results[period] = {'mse': mse, 'rmse': rmse, 'mae': mae, 'r2': r2}

        # actual, predicted prices for each dataset
        for actual, predicted in zip(actual_prices, predicted_prices):
            all_results.append({'period': period, 'actual_prices': actual, 'predicted_prices': predicted})

    # create df with results of actual and predicted prices
    results_df = pd.DataFrame(all_results)

    # display metrics results
    for period, metrics in metrics_results.items():
        print(f"results for {period}:\n mse = {metrics['mse']:.2f}, rmse = {metrics['rmse']:.2f}, mae = {metrics['mae']:.2f}, r2 = {metrics['r2']:.2f}\n")

    return results_df, metrics_results
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def plot_train_val_loss(df, zoom=None):
    """
    Plots training and validation loss for each combination of learning rate and batch size found in the DataFrame.
    Optionally zooms into a specific y-axis range for all plots.

    Parameters:
    - df: DataFrame containing the training data, expected to have multi-level columns.
    - zoom: Optional tuple specifying the (min, max) y-axis limits for zooming into specific loss values.
    """
    # extract unique learning rates and batch sizes
    learning_rates = df[('Hyperparameters', 'Learning Rate')].unique()
    batch_sizes = df[('Hyperparameters', 'Batch Size')].unique()

    # number of subplots needed
    num_plots = len(learning_rates) * len(batch_sizes)
    fig, axs = plt.subplots(num_plots, 1, figsize=(10, 4 * num_plots), squeeze=False)
    plt.style.use('seaborn-whitegrid')

    # flatten axs
    axs = axs.flatten()

    # counter
    plot_idx = 0

    for lr in learning_rates:
        for bs in batch_sizes:
            # filter df for the current learning rate and batch size
            filtered_df = df[(df[('Hyperparameters', 'Learning Rate')] == lr) &
                             (df[('Hyperparameters', 'Batch Size')] == bs)]

            # extract epoch train val loss
            epochs = filtered_df[('Metrics', 'Epoch')]
            training_loss = filtered_df[('Metrics', 'Training Loss')]
            validation_loss = filtered_df[('Metrics', 'Validation Loss')]

            # plot train val loss
            axs[plot_idx].plot(epochs, training_loss, label="Training Loss", marker='o', linestyle='-', color='black')
            axs[plot_idx].plot(epochs, validation_loss, label="Validation Loss", marker='x', linestyle='--', color='black')

            # title labels
            axs[plot_idx].set_title(f"Learning Rate: {lr}, Batch Size: {bs}", fontsize=plot_text_size)
            axs[plot_idx].set_xlabel("Epoch", fontsize=plot_text_size)
            axs[plot_idx].set_ylabel("Loss", fontsize=plot_text_size)

            # zoom
            if zoom is not None:
                axs[plot_idx].set_ylim(zoom[0], zoom[1])

            axs[plot_idx].legend(frameon=True, fontsize=plot_text_size)
            axs[plot_idx].grid(True, which='both', linestyle='dotted', linewidth='0.5')

            # plot counter
            plot_idx += 1

    plt.tight_layout()

    # Show plot
    plt.show()
```

## Actual vs Predicted Plot

```python
def plot_actual_vs_predicted(df, x_range=None, y_range=None):
    """
    Plots actual vs. predicted prices from a DataFrame with optional zooming.

    Parameters:
    - df: DataFrame containing 'Actual Prices' and 'Predicted Prices'.
    - x_range: Tuple (min, max), the range of values to display on the x-axis.
    - y_range: Tuple (min, max), the range of values to display on the y-axis.
    """
    plt.style.use('seaborn-whitegrid')
    plt.figure(figsize=(10, 6))
    plt.scatter(df['Actual Prices'], df['Predicted Prices'], alpha=0.5, color='grey', label='Predicted Prices')

    plt.title('Actual vs. Predicted Prices', fontsize=plot_text_size)
    plt.xlabel('Actual Prices', fontsize=plot_text_size)
    plt.ylabel('Predicted Prices', fontsize=plot_text_size)
    plt.grid(linestyle='dotted')

    if x_range is not None:
        plt.xlim(x_range)
    if y_range is not None:
        plt.ylim(y_range)

    min_line = max(x_range[0], y_range[0]) if x_range and y_range else 0
    max_line = min(x_range[1], y_range[1]) if x_range and y_range else max(df['Actual Prices'].max(), df['Predicted Prices'].max())
    plt.plot([min_line, max_line], [min_line, max_line], '--', color='black', label='Perfect Prediction')

    plt.legend(frameon=True, fontsize=plot_text_size)
    plt.show()
```

# Model Architecture

## CNN1 (Surface + Vol)

```python
import torch
import torch.nn as nn

class HybridModel(nn.Module):
    def __init__(self):
        super(HybridModel, self).__init__()
        # Convolution and pooling layers
        self.conv1 = nn.Conv2d(3, 6, 5, stride=1, padding=0)
        self.pool1 = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5, stride=1, padding=0)
        self.pool2 = nn.MaxPool2d(2, 2)
        self.conv3 = nn.Conv2d(16, 32, 5, stride=1, padding=2)
        self.pool3 = nn.MaxPool2d(2, 2)

        self.fc_image = nn.Linear(32 * 6 * 6, 200)

        # Fully connected layers
        self.fc1 = nn.Linear(200 + 5, 100)
        self.fc2 = nn.Linear(100, 50)
        self.fc_out = nn.Linear(50, 1)

        self.relu = nn.ReLU(inplace=True)
        self.input_dropout = nn.Dropout(0.5)
        self.dropout = nn.Dropout(0.5)

    def forward(self, inputs):
        image, additional_features = inputs

        x = self.pool1(self.relu(self.conv1(image)))
        x = self.pool2(self.relu(self.conv2(x)))
        x = self.pool3(self.relu(self.conv3(x)))

        # Flatten the output from the last convolutional layer
        x = x.view(-1, 32 * 6 * 6)  #
        x = self.relu(self.fc_image(x))

        # Concatenate image features with additional features
        x = torch.cat((x, additional_features), dim=1)

        # input layer dropout
        x = self.input_dropout(x)
        # relu
        x = self.relu(self.fc1(x))
        #dropout
        x = self.dropout(x)

        x = self.relu(self.fc2(x))

        x = self.fc_out(x)

        return x
```

```python
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import pandas as pd
import os
import torch

class VolatilitySurfaceDataset(Dataset):
    def __init__(self, images_dir, dataframe, transform=None):
        # ensure log_price column exists in the df
        assert 'log_price' in dataframe.columns
        self.images_dir = images_dir
        self.dataframe = dataframe
        self.transform = transform

    def __len__(self):
        # return the number of samples in the dataset
        return len(self.dataframe)

    def __getitem__(self, idx):
        # get the formatted date for the image file name
        formatted_date = self.dataframe.iloc[idx]['date'].strftime('%Y-%m-%d')
        # construct the image file path
        img_name = os.path.join(self.images_dir, f"vol_surface_{formatted_date}.png")
        # open the image and convert to RGB
        image = Image.open(img_name).convert('RGB')
        # apply transformations if any
        if self.transform:
            image = self.transform(image)

        # extract additional features
        strike_price = self.dataframe.iloc[idx]['log_strike_price']
        maturity = self.dataframe.iloc[idx]['maturity']
        volatility = self.dataframe.iloc[idx]['Vola_ATM']
        spindx = self.dataframe.iloc[idx]['log_spindx']
        t30ret = self.dataframe.iloc[idx]['t30ret']

        # convert features to tensor
        additional_features = torch.tensor([strike_price, maturity, volatility, spindx, t30ret], dtype=torch.float32)

        # extract and convert target to tensor
        target = self.dataframe.iloc[idx]['log_price']
        target = torch.tensor(target, dtype=torch.float32)

        return (image, additional_features), target
```

```python
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import pandas as pd
import os
import torch

class VolatilitySurfaceDataset(Dataset):
    def __init__(self, images_dir, dataframe, transform=None):
        # ensure log_price column exists in df
        assert 'log_price' in dataframe.columns
        self.images_dir = images_dir
        self.dataframe = dataframe
        self.transform = transform

    def __len__(self):
        # return the number of samples in the data
        return len(self.dataframe)

    def __getitem__(self, idx):
        # get the formatted date for the image file name
        formatted_date = self.dataframe.iloc[idx]['date'].strftime('%Y-%m-%d')
        # construct the image file path
        img_name = os.path.join(self.images_dir, f"vol_surface_{formatted_date}.png")
        # open the image and convert to RGB
        image = Image.open(img_name).convert('RGB')
        # apply transformations if any
        if self.transform:
            image = self.transform(image)

        # extract additional features
        strike_price = self.dataframe.iloc[idx]['log_strike_price']
        maturity = self.dataframe.iloc[idx]['maturity']
        spindx = self.dataframe.iloc[idx]['log_spindx']
        t30ret = self.dataframe.iloc[idx]['t30ret']

        # convert features to tensor
        additional_features = torch.tensor([strike_price, maturity, spindx, t30ret], dtype=torch.float32)

        # extract and convert target to tensor
        target = self.dataframe.iloc[idx]['log_price']
        target = torch.tensor(target, dtype=torch.float32)

        return (image, additional_features), target
```
