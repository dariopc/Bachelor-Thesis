## Volatility Skew

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# filter df for the year 2020
df_filtered = bsm_df[(bsm_df['date'].dt.year) == 2020]

# calculate  average underlying price
underlying_price = df_filtered['spindx'].mean()

# group by strike_price and calculate the average of impl_volatility for each group
df_grouped = df_filtered.groupby('strike_price')['impl_volatility'].mean().reset_index()

# sort the grouped df by strike price
df_sorted = df_grouped.sort_values(by='strike_price')

# setting the visual style of seaborn plots
sns.set(style='whitegrid')

# create a new plot with defined dimensions
plt.figure(figsize=(10, 6))

# plot the average implied volatility against strike price
plt.plot(df_sorted['strike_price'], df_sorted['impl_volatility'], linestyle='-', color='black')

# mark the price of the underlying on the x-axis
plt.axvline(x=underlying_price, color='black', linestyle='--', label=f'Ã˜ Underlying Price ({underlying_price:.2f})')

# set label grid legend
plt.xlabel('Strike Price', fontsize=plot_text_size)
plt.ylabel('Average Implied Volatility', fontsize=plot_text_size)
plt.legend(frameon=True, fontsize=plot_text_size)
plt.grid(linestyle='dotted')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# datetime
bsm_df['date'] = pd.to_datetime(bsm_df['date'])

# initialize the figure and set visual style
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))

# define years to plot and their respective colors and line styles
years = [2020, 2021, 2022]
colors = ['black', '#646464', '#AAAAAA']
line_styles = [':', '--', '-.']

for year, color, line_style in zip(years, colors, line_styles):
    # filter data for the specific year
    df_filtered = bsm_df[bsm_df['date'].dt.year == year]

    # calculate the mean underlying price for the year
    underlying_price = df_filtered['spindx'].mean()

    # group by strike_price and calculate the average of impl_volatility for each group
    df_grouped = df_filtered.groupby('strike_price')['impl_volatility'].mean().reset_index()

    # sort the grouped df by strike price
    df_sorted = df_grouped.sort_values(by='strike_price')

    # plot the volatility skew for the year
    plt.plot(df_sorted['strike_price'], df_sorted['impl_volatility'], linestyle='-', label=f'{year}', color=color)

    # mark the average price of the underlying for the year
    plt.axvline(x=underlying_price, color=color, linestyle=line_style, label=f'Underlying Price')

# setting plot labels and title
plt.xlabel('Strike Price', fontsize=plot_text_size)
plt.ylabel('Average Implied Volatility', fontsize=plot_text_size)
plt.legend(frameon=True, loc='upper right', fontsize=plot_text_size)
plt.grid(True, linestyle='dotted')

# save plot
plt.savefig(path+'volatility_skew', dpi=500)
plt.show()
```

```python
import scipy.stats as stats
import math

def black_scholes(S, K, T, r, sigma, option='call'):
    """
    black_scholes calculates the price of a call or put option using the Black-Scholes formula.

    :param S: Current stock price
    :param K: Option strike price
    :param T: Time to expiration in years
    :param r: Risk-free interest rate
    :param sigma: Volatility of the stock
    :param option: Type of option ('call' or 'put')
    :return: Price of the option
    """
    # calculate d1 and d2 for the option's price
    d1 = (math.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * math.sqrt(T))
    d2 = (math.log(S / K) + (r - 0.5 * sigma ** 2) * T) / (sigma * math.sqrt(T))

    # compute and return the option's price based on option type
    if option == 'call':
        return S * stats.norm.cdf(d1) - K * math.exp(-r * T) * stats.norm.cdf(d2)
    elif option == 'put':
        return K * math.exp(-r * T) * stats.norm.cdf(-d2) - S * stats.norm.cdf(-d1)

    # error handling
    else:
        raise ValueError("Invalid option type. Use 'call' or 'put'.")
```

```python
import pandas as pd

def find_atm_calls(df):
    """
    find_atm_calls looks for a call with the delta closest to 0.5.

    :param df: input dafaframe
    :return: pandas Series representing the row that is closest to 0.5
    """
    # calculate difference delta and 0.5
    abs_delta_diff = df['delta'].sub(0.5).abs()

    # look for index
    return df.loc[abs_delta_diff.idxmin()]
```

```python
# group by date and apply find_atm_calls function to find at-the-money calls
atm_calls = bsm_df.groupby('date', as_index=False).apply(find_atm_calls).reset_index(drop=True)

# calculate daily average implied volatility for at-the-money calls
daily_vola_atm = atm_calls.groupby('date')['impl_volatility'].mean().reset_index()

# rename column to Vola_ATM
daily_vola_atm.rename(columns={'impl_volatility': 'Vola_ATM'}, inplace=True)

# merge daily average implied volatility with the original dataframe
bsm_df = pd.merge(bsm_df, daily_vola_atm, on='date', how='left')
```

```python
# initialize an empty list to store BSM values for ATM options
bsm_values_ATM = []

# iterate over each row in the dataframe
for index, row in bsm_df.iterrows():
    # calculate the Black-Scholes value using Vola_ATM for each row
    bsm_value = black_scholes(row["spindx"], row["strike_price"], row["maturity"], row["t30ret"], row["Vola_ATM"], 'call')
    # append the calculated value to the list
    bsm_values_ATM.append(bsm_value)

# add the calculated BSM values as a new column to the dataframe
bsm_df["BSM_Vola_ATM"] = bsm_values_ATM
```
